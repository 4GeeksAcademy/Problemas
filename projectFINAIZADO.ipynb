```python
# TODO
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.preprocessing import MinMaxScaler
ds = pd.read_csv('Proyecto_prework/real_estate.csv', sep=";")

poblaciones = ['Fuenlabrada', 'Leganés', 'Getafe', 'Alcorcón']
subconjunto = ds[ds['level5'].isin(poblaciones)]

# Calcula el precio por metro cuadrado (pps)
subconjunto['pps'] = subconjunto['price'] / subconjunto['surface']

# Filtra solo las localidades de "Getafe" y "Alcorcón"
getafe_alcorcon = subconjunto[subconjunto['level5'].isin(['Getafe', 'Alcorcón'])]

# Imprime información estadística sobre el precio por metro cuadrado
print(getafe_alcorcon.groupby('level5')['pps'].describe())
```

#### Exercise 20. Make the same plot for 4 different populations (level5 column) and rearrange them on the same graph? You must use the subset obtained in the question 14 (★★☆) 
Hint: make a scatter plot of each population using subplots.

```python
# TODO
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.preprocessing import MinMaxScaler
ds = pd.read_csv('Proyecto_prework/real_estate.csv', sep=";")

poblaciones = ['Fuenlabrada', 'Leganés', 'Getafe', 'Alcorcón']
subconjunto = ds[ds['level5'].isin(poblaciones)]

# Calcula el precio por metro cuadrado (pps)
subconjunto['pps'] = subconjunto['price'] / subconjunto['surface']

# Crea un gráfico de dispersión para cada población usando subgráficos
poblaciones = ['Fuenlabrada', 'Leganés', 'Getafe', 'Alcorcón']
fig, axes = plt.subplots(nrows=2, ncols=2, figsize=(12, 10))

for i, poblacion in enumerate(poblaciones):
    datos_poblacion = subconjunto[subconjunto['level5'] == poblacion]
    axes[i // 2, i % 2].scatter(datos_poblacion['surface'], datos_poblacion['pps'], label=poblacion, alpha=0.7)
    axes[i // 2, i % 2].set_title(f'Gráfico de Dispersión - {poblacion}')
    axes[i // 2, i % 2].set_xlabel('Metros Cuadrados')
    axes[i // 2, i % 2].set_ylabel('Precio por Metro Cuadrado')
    axes[i // 2, i % 2].legend()

plt.tight_layout()
plt.show()
```
